# Feature Engineering

Feature engineering includes a range of multiple engineering techniques
such as selecting relevant features, handling missing data, encoding the data,
normalizing it. It is one of the most crucial tasks and plays a major role 
in determining the outcome of a model 

```python
sales
```

```python
sales["Id_Item_Cat"]=sales.Item_Identifier.str[0:2]
```

```python
sales.Item_Fat_Content.unique()
```

```python
# We need to change the category of NC into a new category 
```

```python
sales.loc[sales.Id_Item_Cat=='NC','Item_Fat_Content']='Non Edible'
```

```python
# Creating categories perishable and non persihable
```

```python
perish=['Dairy', 'Meat', 'Fruits and Vegetables','Breakfast',  
       'Breads', 'Starchy Foods',  'Seafood']
```

```python
def perishkardo(x):
    if x in perish:
        return('Perishable')
    else:
        return('Non Perishable')
```

```python
sales['Item_cat']=sales.Item_Type.apply(perishkardo)
```

```python
# Lets talk about outlet Id
# binning the outlet id into three categories
```

```python
lowes =["OUT010", "OUT019"]
high =["OUT013", "OUT027"]


def outlets(x):
    if x in lowes:
        return("Low_performers")
    elif x in high:
        return("high_performers")
    else: 
        return("Avg_performers") 
    
    
sales["Out_Cat"]=sales.Outlet_Identifier.apply(outlets)
```

```python
# establishment Year
sales['vintage']=2013-sales.Outlet_Establishment_Year
```

# Missing Values

Missing values/or null values are basically void values, which could have 
been entered because of any human error or some technical glitch in the 
dataset, and thus we need to trat them before moving into preprocessing 
stage as this may hinder our process of making the most efficient model

```python
sales.loc[sales.Item_Identifier=='FDA15'] # Item weight has the missing values and it is dependent on Item identifier
```

```python
# so we will take the Item Identifier to replace the missing value 
```

```python
# There are two functions in fillna()
# 1.fflill() # forward row filling
# 2.bfill()  # backward row filling 
```

```python
sales['Item_Weight']=sales.groupby('Item_Identifier')['Item_Weight'].transform(lambda x:x.fillna(x.mean()))
```

```python
sales.loc[sales.Item_Weight.isnull()] # These are the single entries so the mean is not there 
# so the fill na method did not work 
```

# SCALING:

Scaling or Standardization is the essential step of data preprocessing
that is applied to independent variables or features of data.
It is used to normalize that data upto a certain range, and thus 
helps in speeding up the calculations in an algorithm.

It is used with the Python Package: sklearn.preprocessing.

```python
sales
```

```python
sales.columns
```

```python
# Now we would be dropping the unnecessary variables. 
```

```python
Finaldata=sales.drop(['Item_Identifier', 
       'Item_Type', 'Outlet_Identifier',
       'Outlet_Establishment_Year'],axis=1)
```

```python
Finaldata
```

```python
##Importing scaler package for scaling process.

from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
Finaldata['scaled_sales']=sc.fit_transform(pd.DataFrame(Finaldata.Item_Outlet_Sales))
```

```python
sc=StandardScaler()
Finaldata['scaled_mrp']=sc.fit_transform(pd.DataFrame(Finaldata.Item_MRP))
```

```python
sc=StandardScaler()
Finaldata["scaled_Weight"]=sc.fit_transform(pd.DataFrame(Finaldata.Item_Weight))
```

```python
sc=StandardScaler()
Finaldata["scaled_visibile"]=sc.fit_transform(pd.DataFrame(Finaldata.Item_Visibility))
```

```python
sc=StandardScaler()
Finaldata["scaled_vintage"]=sc.fit_transform(pd.DataFrame(Finaldata.vintage))
```
